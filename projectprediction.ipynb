```python
import pandas as pd
import numpy as np
```

```python
matches=pd.read_csv("ipl-matches.csv")
ball=pd.read_csv("IPL_Ball_by_Ball_2008_2022.csv")
```

```python
matches_detail=matches.merge(ball,on="ID")
```

```python
matches_detail["Team1"]=matches_detail["Team1"].str.replace("Deccan Chargers","Sunrisers Hyderabad")
matches_detail["Team2"]=matches_detail["Team2"].str.replace("Deccan Chargers","Sunrisers Hyderabad")
matches_detail["Team1"]=matches_detail["Team1"].str.replace("Delhi Daredevils","Delhi Capital")
matches_detail["Team2"]=matches_detail["Team2"].str.replace("Delhi Daredevils","Delhi Capital")
matches_detail["Team1"]=matches_detail["Team1"].str.replace("Kings XI Punjab","Punjab Kings")
matches_detail["Team2"]=matches_detail["Team2"].str.replace("Kings XI Punjab","Punjab Kings")
```

```python
matches_detail.iloc[1]
```

```python
matches
```

```python
team=["Chennai Super Kings","Delhi Capitals","Gujarat Titans","Punjab Kings","Lucknow Super Giants","Mumbai Indians","Kolkata Knight Riders","Rajasthan Royals","Royal Challengers Bangalore","Sunrisers Hyderabad"]                                   
  
```

```python
matches_detail=matches_detail[(matches_detail["Team1"].isin(team)) & (matches_detail["Team2"].isin(team) )]
```

```python
total_score=matches_detail.groupby(["ID","innings"])["total_run"].sum()
total_score=total_score.reset_index()
ts=total_score[total_score["innings"]==1]
```

```python
matches_detail=matches_detail[matches_detail["innings"]==2]
```

```python
matches_detail=matches_detail.merge(ts,on="ID")
```

```python
matches_detail["runs"]=matches_detail.groupby("ID")["total_run_x"].cumsum()
```

```python
matches_detail["runs_left"]=matches_detail["total_run_y"]-matches_detail["runs"]
```

```python
matches_detail=matches_detail[matches_detail["runs_left"]>=0]
```

```python
def result(row):
    return 1 if row["BattingTeam"]==row["WinningTeam"] else 0
```

```python
matches_detail["result"]=matches_detail.apply(result,axis=1)
```

```python
matches_detail.sample()
```

```python
matches_detail["wickets_left"]=10-matches_detail.groupby("ID")["isWicketDelivery"].cumsum()
```

```python
matches_detail[matches_detail["innings_x"]==1]
```

```python
ball
```

```python
matches_detail["balls_left"]=120-(matches_detail["overs"]*6+matches_detail["ballnumber"])
```

```python
matches_detail=matches_detail[matches_detail["balls_left"]!=0]
```

```python
matches_detail["CRR"]=matches_detail["runs"]*6/(120-matches_detail["balls_left"])
```

```python
matches_detail["RRR"]=matches_detail["runs_left"]*6/(matches_detail["balls_left"])
```

```python
def bowl(row):
    if row["BattingTeam"]!=row["Team1"]:
        return row["Team1"]
    else:
        return row["Team2"]
        
```

```python
matches_detail["BowlingTeam"]=matches_detail.apply(bowl,axis=1)
```

```python
matches_detail.iloc[170]
```

```python
final=matches_detail[["BattingTeam","BowlingTeam","City","runs_left","total_run_y","balls_left","wickets_left","CRR","RRR","result"]]
```

```python
final
```

```python
x=final.iloc[:,:-1]
y=final.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf=ColumnTransformer([
    ("trf",OneHotEncoder(sparse=False,drop="first"),["BattingTeam","BowlingTeam","City"])
],remainder="passthrough")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe=Pipeline(steps=[
    ("step1",trf),
    ("step2",LogisticRegression(solver="liblinear"))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
